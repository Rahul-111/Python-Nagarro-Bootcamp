```python
class Circle:
    def __init__(self,radius):
        self.radius=radius
    def perimeter(self):
        print(2*3.14*self.radius)
    def area(self):
        print(3.14*self.radius**2)
    def __lt__(self, other):# For x < y
        return self.radius < other.radius 
    
    def __gt__(self, other): # For x > y
        return self.radius > other.radius
    
    def __le__(self, other): # For x <= y
        return self.radius <= other.radius
    
    def __ge__(self, other): # For x >= y
        return self.radius >= other.radius
    
    def __eq__(self, other): # For x == y
        return self.radius == other.radius
    
    def __ne__(self, other):# For x != y OR x <> y
        return self.radius != other.radius
    
```

```python
c1=Circle(12)
c2=Circle(10)
```

```python
if c1<c2:
    print(str(c1.radius)+' is smaller then ' +str(c2.radius))
elif c1>c2:
    print(str(c1.radius)+' is greater then ' +str(c2.radius))
elif c1<=c2:
    print(str(c1.radius)+' is less then equal or equal to ' +str(c2.radius))
elif c1>=c2:
    print(str(c1.radius)+' is greater then or equal to ' +str(c2.radius))
elif c1==c2:
    print(str(c1.radius)+' is equal ' +str(c2.radius)) 
elif c1!=c2:
    print(str(c1.radius)+' is not equal to ' +str(c2.radius))
```

```python
with open('textfile.txt' ,'r') as f:
        l=f.read()
```

```python
l
```

```python
listOfWord=[]
l=l.split(' ')
for word in l:
    listOfWord.append(word)
```

```python
listOfWord
```

```python
list(map((lambda x:x.upper()),listOfWord))
```

```python
finalList=list(filter(lambda x:x.find('a')==-1,listOfWord))
```

```python
ans=''
for word in finalList:
    ans+=word
```
