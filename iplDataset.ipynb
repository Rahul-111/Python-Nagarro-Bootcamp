```python
df=pd.read_csv('matches.csv')
df.head()
```

```python
team1=df['team1']
team2=df['team2']
winner=df['winner']
```

```python
losser=[]
for i in range(len(team1)):
    if winner[i]==team1[i]:
        losser.append(team2[i])
    elif winner[i]==team2[i]:
        losser.append(team1[i])
    else:
        losser.append(0)#because some time match get draw
```

```python
len(losser)
```

```python
len(team1)
df['Losser']=losser
```

```python
df.head()
```

```python
df[df['win_by_wickets']==0]['Losser'].value_counts()
```
